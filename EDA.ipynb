```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ruta del archivo armadores.xlsx
ruta_armadores = './data_inicial/armadores/armadores.xlsx'

# Ruta del archivo repartidores.xlsx
ruta_repartidores = './data_inicial/repartidores/repartidores.xlsx'

# Leer el archivo armadores.xlsx y almacenarlo en un DataFrame llamado df_armadores
df_armadores = pd.read_excel(ruta_armadores)

# Leer el archivo repartidores.xlsx y almacenarlo en un DataFrame llamado df_repartidores
df_repartidores = pd.read_excel(ruta_repartidores)

```

## EDA Repatidores


```python
df_repartidores.head(2)
```

```python
df_repartidores.describe()
```

```python
# Gráficos para 'tarde'
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df_repartidores['tarde'], kde=True)
plt.title('Distribución de Tardanzas')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_repartidores['tarde'])
plt.title('Boxplot de Tardanzas')

plt.show()
```

```python
# Gráficos para 'completados'
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df_repartidores['completados'], kde=True)
plt.title('Distribución de Pedidos Completadas')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_repartidores['completados'])
plt.title('Boxplot de Pedidos Completados')

plt.show()
```
